# **Titanic Survival Analysis (Google Sheets)**

## **Objective**

The primary objective of this case study is to analyze the survival rates of passengers on the Titanic, focusing on different demographic factors such as gender, passenger class, and age groups. This analysis aims to uncover significant patterns and insights that influenced survival outcomes during the disaster.


## **Key Questions**

1. What was the overall survival rate of passengers on the Titanic?
2. How did survival rates vary between male and female passengers?
3. What were the survival rates for passengers in different classes (1st, 2nd, 3rd)?
4. How did the survival rates compare between children (under 18 years) and adults?


## **Steps Taken**

### **1. Finding the Data:**
The Titanic dataset was sourced from Kaggle, which includes information on passengers, their demographic details, and whether they survived the disaster.
	
### **2. Preparing and Cleaning the Data:**
Import Data: The dataset was imported into Google Sheets for analysis.
Handling Missing Values: Missing ages were filled with the mean age (30) to ensure completeness of data. It’s important to note that this may introduce some inaccuracies.
Creating Age Groups: A new column was added to categorize passengers into age groups (Children and Adults).

### **3. Data Analysis:**
Overall Survival Rate: Calculated the total number of survivors and non-survivors to determine the overall survival rate.
Survival Rate by Gender: Created pivot tables to calculate survival rates for male and female passengers.
Survival Rate by Passenger Class: Analyzed survival rates across different passenger classes.
Survival Rate by Age Group: Compared survival rates between children (under 18) and adults.

### **4. Visualization:**
Various charts were created to visually represent the findings, including pie charts and bar charts.

## **Key Insights**
	
### **Overall Survival Rate:**
38.4% of passengers survived, while 61.6% did not survive.

### **Gender Disparity in Survival Rates:**
Female passengers had a significantly higher survival rate (74.2%) compared to male passengers (18.9%).
The majority of the survivors were female (79.7%).

### **Class Disparity in Survival Rates:**
First-class passengers had the highest survival rate (63.0%), followed by second-class passengers (47.3%) and third-class passengers (24.2%).

### **Age Disparity in Survival Rates:**
Children had a higher survival rate (54.0%) compared to adults (36.1%), indicating a better chance of survival for children.
Note: The accuracy of this metric is limited due to the use of mean age for missing data.


## **Conclusion**

The analysis reveals significant disparities in survival rates on the Titanic based on gender, passenger class, and age group. Female passengers and children had higher chances of survival compared to male passengers and adults. First-class passengers also had a higher survival rate compared to those in second and third classes. These insights highlight the importance of demographic factors in survival outcomes during the Titanic disaster.
