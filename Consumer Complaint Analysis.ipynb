## **#1 Top 10 Complaint Issues**





### **Introduction**

This report analyzes the top 10 complaint issues reported by consumers in the CFPB (Consumer Financial Protection Bureau) complaints database. The analysis provides insights into the most frequent issues faced by consumers across various financial products and services.

### **Data Source**

The data is sourced from the CFPB complaints database, which records consumer complaints about financial products and services. The analysis focuses on the top 10 issues by complaint count, along with their respective percentages of the total complaints.

### **Methodology**

The analysis is conducted using SQL queries on the CFPB complaints database hosted on BigQuery. The query used for this analysis is:

width="3000px">

### **Results**

The total number of complaints recorded in the CFPB complaints database is 3,458,906.

The code below shows the top 10 complaint issues by count, including the percentage of total complaints they represent:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
issues = ["Incorrect information on your report", "Problem with a credit reporting company's investigation", 
          "Improper use of your report", "Attempts to collect debt not owed", 
          "Loan modification, collection, foreclosure", "Incorrect information on credit report", 
          "Managing an account", "Loan servicing, payments, escrow account", 
          "Written notification about debt", "Trouble during payment process"]
counts = [858157, 456973, 372875, 164028, 112307, 102684, 98447, 77332, 73216, 68255]
percentages = [24.81, 13.21, 10.78, 4.74, 3.25, 2.97, 2.85, 2.24, 2.12, 1.97]

# Plotting
plt.figure(figsize=(14, 8))
bars = plt.barh(issues, counts, color='skyblue')
plt.xlabel('Number of Complaints')
plt.title('Top 10 Complaint Issues by Count (with Percentages)')
plt.gca().invert_yaxis()

# Adding percentage labels to each bar
for bar, percentage in zip(bars, percentages):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{percentage}%', ha='left', va='center')

plt.show()
```

In addition to the bar chart, the pie chart below shows the distribution of the top 3 issues compared to all other issues:

```python
import matplotlib.pyplot as plt

# Data
issues = ["Incorrect information on your report", "Problem with a credit reporting company's investigation", 
          "Improper use of your report", "Other"]
counts = [858157, 456973, 372875, 1820598]  # 1,820,598 represents "Other" companies
percentages = [24.81, 13.21, 10.78, 51.20] 

# Colors for the pie chart
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Plotting
plt.figure(figsize=(10, 7))
plt.pie(counts, labels=issues, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Top 3 Issues vs Other in Complaint Counts (with Percentages)')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

plt.show()
```

### **Focus on Top 3 Issues**

**1. Incorrect Information on Your Report:**
* Complaint Count: 858,157
* Percentage of Total Complaints: 24.81%
* This issue tops the list with the highest number of complaints, indicating a major concern for consumers regarding the accuracy of information in their credit reports.

**2. Problem with a Credit Reporting Company’s Investigation:**
* Complaint Count: 456,973
* Percentage of Total Complaints: 13.21%
* Consumers face significant challenges when dealing with credit reporting companies’ investigations into disputes, highlighting issues with the responsiveness and effectiveness of these investigations.

**3. Improper Use of Your Report:**
* Complaint Count: 372,875
* Percentage of Total Complaints: 10.78%
* This issue reflects concerns over the misuse of consumer credit reports, including unauthorized access and errors in report usage.

### **Conclusion**

The analysis identifies the top 10 complaint issues in the CFPB complaints database, with a focus on the top 3 issues. The results show that incorrect information on credit reports is the most common complaint, followed by problems with credit reporting company investigations and improper use of credit reports. These insights highlight the need for improved accuracy, transparency, and responsiveness in credit reporting processes.

In addition to the bar chart, the pie chart below shows the top 3 companies by complaint count compared to an “Other” category, illustrating the proportion of complaints for each:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
companies = ["EQUIFAX, INC.", "TRANSUNION INTERMEDIATE HOLDINGS, INC.", 
             "Experian Information Solutions, Inc.", "Other"]
counts = [619370, 526785, 492153, 1820598]  # 1,820,598 represents "Other" companies
percentages = [17.91, 15.23, 14.23, 52.63]  

# Colors for the pie chart
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Plotting
plt.figure(figsize=(10, 7))
plt.pie(counts, labels=companies, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Top 3 Companies vs Other in Complaint Counts (with Percentages)')
plt.axis('equal')

plt.show()
```

### **Observations and Patterns**

**2. Major Financial Institutions:**
* Large banks such as Bank of America, Wells Fargo, and JPMorgan Chase are also among the top companies with high complaint counts.
* This reflects widespread consumer interactions with these institutions and the potential for various service-related issues.

**3. Consumer Financial Services:**
* Companies providing financial services, such as Capital One and Synchrony Financial, also appear in the top 10.
* These companies face a considerable number of consumer complaints, highlighting areas for service improvements.
    
### **Conclusion**

The analysis identifies the top 10 companies by complaint count in the CFPB complaints database, providing insights into where most consumer complaints are concentrated. Credit reporting agencies account for nearly half of the total complaints, indicating significant consumer concerns in this area. Large financial institutions and financial service providers also have substantial complaint counts, suggesting opportunities for improving customer service and addressing consumer issues more effectively.
